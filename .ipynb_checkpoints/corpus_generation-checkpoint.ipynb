# Generate queries and target actions across categories

```python
categories = {
    "Computer Usage and Accessibility": [
            "Assist with text rephrasing, corrections, and clarity improvements",
            "Speech-based web searches",
            "Automatic page summarization",
            "Navigation between links and menus",
            "Writing, replying to, and summarizing emails",
            "Searching for files by keywords or topics",
            "Enable seamless navigation and multi-app coordination.",
            "File opening and saving",
            "Multi-app coordination (e.g., copy-pasting between apps)",
            "Composing and reading messages in apps like WhatsApp or iMessage",
            "Creating posts, commenting, and managing accounts on social media",
            "Scheduling meetings and sending invites",
            "Support online banking activities.",
            "Assist with browsing and purchasing items online.",
            "Ordering groceries online or scheduling deliveries",
            "Booking rides via Uber, Lyft, etc.",
            "Provide public transit schedules and guidance.",
            "Help fill out online forms or documents.",
    ],
    "Smart Home and Environment Control": [
            "Controlling lights, thermostats, and fans",
            "Adjusting curtains or blinds",
            "Operate home appliances remotely.",
            "Starting/stopping appliances (e.g., dishwasher, laundry)",
            "Manage home security systems and monitor cameras.",
            "Managing door locks, cameras, and alarms",
            "Answering video doorbells",
            "Access and control entertainment systems.",
            "Controlling TVs, speakers, or streaming services",
            "Searching for and playing specific content"
    ],
    "Healthcare and Wellness": [
            "Create reminders for medication schedules",
            "Create reminders for appointments",
            "Summoning help in emergencies",
            "Contacting caretakers or family members",
            "Querying for predefined responses to communicate caregiving needs"
    ]
}
```

```python
# define model and prompt
model = "gpt-4o-mini"
n_gens = 1
system_prompt = (
    "You are a helpful assistant that generates concise user queries and corresponding target actions for an LLM-agent framework. "
    "This framework is designed to assist individuals with paralysis. "
    "Your task is to generate outputs in the following JSON format for each query-target pair:\n\n"
    "{\n"
    '  "query": "<User query (â‰¤ 5 words)>",\n'
    '  "target": "<Target action description>"\n'
    "}\n\n"
    "Ensure the query is no more than three words long and the target is detailed yet concise."
)

# generate queries and target outcomes
completions = {}
for cat, ideas in categories.items():
    print(f"Category == {cat}")
    completions[cat] = {}
    for idea in ideas:
        completions[cat][idea] = []
        for n in range(n_gens):
            completion = client.chat.completions.create(
                model=model,
                messages=[
                    {"role": "system", "content": system_prompt},
                    {
                        "role": "user",
                        "content": (
                            f"Category: {cat}\n"
                            f"Subcategory: {idea}\n\n"
                            "Generate a query and target action based on the category and subcategory above."
                        )
                    }
                ]
            )
            print(completion.choices[0].message.content)
            completions[cat][idea].append(completion)
```

```python
# save prompt into corpus directory. comment out to execute
raise

import os
import pickle as pk
import json

corpus_path = f'corpus/{model}'
if not os.path.exists(corpus_path):
    os.makedirs(corpus_path)

# save all completions into a pickle file
pickle_file_path = os.path.join(corpus_path, 'completions.pkl')
with open(pickle_file_path, 'wb') as pkl_file:
    pk.dump(completions, pkl_file)

# save completions into a text file line by line
text_file_path = os.path.join(corpus_path, 'completions.txt')
delimiter = " | " 
header = (
    "This file contains user query and target action pairs generated by an LLM-agent framework.\n"
    "Each line consists of: Category, Subcategory, Query, Target.\n"
    f"Values are separated by '{delimiter}'.\n"
    "\n"
)
with open(text_file_path, 'w') as txt_file:
    txt_file.write(header)
    for cat, ideas in completions.items():
        for idea, outputs in ideas.items():
            for output in outputs:
                content = output.choices[0].message.content
                content = json.loads(content)
                query = content['query']
                target = content['target']
                txt_file.write(f"{cat}{delimiter}{idea}{delimiter}{query}{delimiter}{target}\n")
print("completions saved")
```
